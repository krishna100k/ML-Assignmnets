```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df
```

```python
#Understanding the Data
print(df.head())
print(df.describe())
print(df.info())
```

```python
# Univariate Analysis
# Explore each variable individually.
# For numerical features, create histograms or kernel density plots.
# For categorical features, create bar charts.

#Bivariate
# Relationship between the two variables
sns.histplot(data = df, x = "tip")
```

```python
#regPlot
sns.regplot(data = df, x="total_bill", y="tip")
plt.title("Regression Plot of Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
#Heatmap
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, cmap="coolwarm", annot = True)
```
