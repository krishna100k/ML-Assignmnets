```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv")
df
```

```python
x = np.array([df["Annual Income (k$)"], df["Spending Score (1-100)"]])
```

```python
plt.scatter(x[0], x[1])
```

```python
from sklearn.cluster import KMeans as km
```

```python
wcss = []
for i in range (1, 11):
    kmeans = km(i)
    kmeans.fit(x.T)
    w = kmeans.inertia_
    wcss.append(w)
wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans = km(5)
kmeans.fit(x.T)
labels = kmeans.labels_
```

```python
plt.scatter(x[0], x[1], c = labels)
```
